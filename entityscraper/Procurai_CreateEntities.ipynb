```python
from bs4 import BeautifulSoup
import requests
from string import ascii_lowercase
import sys 


req_headers = {
    'accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3',
    'accept-encoding': 'gzip, deflate, br',
    'accept-language': 'en-US,en;q=0.8',
    'cache-control': 'max-age=0',
    'upgrade-insecure-requests': '1',
    'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/61.0.3163.100 Safari/537.36'
}
print(req_headers)
```

```python
symptomurl = 'https://www.healthline.com/api/symptom-checker/symptoms?rows=100&start=0&searchTerms='
print(symptomurl)
```

```python
l = []
for singlealphabet in ascii_lowercase:
    with requests.Session() as s:
        r = s.get(symptomurl+singlealphabet, headers=req_headers)
        soup = BeautifulSoup(r.content, 'lxml')
        a = soup.text
        astr = a.replace("[","").replace("]","").replace('"','')
        li = list(astr.split(",")) 
        l = l + li
```

```python
def getsimplevariants(singlesymptom, finalstring):
    
    if  (singlesymptom.find(' ') == singlesymptom.rfind(' ') and singlesymptom.find(' ') != -1):
        splitstring = singlesymptom.split(' ')
        splitstring[0], splitstring[1] = splitstring[1], splitstring[0]
        finalstring = finalstring + ',"' + ' '.join(splitstring) + '"'
        
    if  (singlesymptom.find(' ') == singlesymptom.rfind(' ') and singlesymptom.find(' ') != -1):
        splitstring = singlesymptom.split(' ')
        splitstring[0], splitstring[1] = splitstring[1], splitstring[0]
        finalstring = finalstring + ',"' + ' of the '.join(splitstring) + '"'
        
    if  (singlesymptom.find(' ') == singlesymptom.rfind(' ') and singlesymptom.find(' ') != -1):
        splitstring = singlesymptom.split(' ')
        splitstring[0], splitstring[1] = splitstring[1], splitstring[0]
        finalstring = finalstring + ',"' + ' in the '.join(splitstring) + '"'
        
    if  (singlesymptom.find(' ') == singlesymptom.rfind(' ') and singlesymptom.find(' ') != -1):
        splitstring = singlesymptom.split(' ')
        splitstring[0], splitstring[1] = splitstring[1], splitstring[0]
        finalstring = finalstring + ',"' + ' in '.join(splitstring) + '"'
        
    if  (singlesymptom.find(' ') == singlesymptom.rfind(' ') and singlesymptom.find(' ') != -1):
        splitstring = singlesymptom.split(' ')
        splitstring[0], splitstring[1] = splitstring[1], splitstring[0]
        finalstring = finalstring + ',"' + ' of '.join(splitstring) + '"'
  
    if  (singlesymptom.find(' ') == singlesymptom.rfind(' ') and singlesymptom.find(' ') != -1):
        finalstring = finalstring + ',"' + singlesymptom.replace(' ', ' in the ') + '"'
        
    if  (singlesymptom.find(' ') == singlesymptom.rfind(' ') and singlesymptom.find(' ') != -1):
        finalstring = finalstring + ',"' + singlesymptom.replace(' ', ' of the ') + '"'
        
    if  (singlesymptom.find(' ') == singlesymptom.rfind(' ') and singlesymptom.find(' ') != -1):
        finalstring = finalstring + ',"' + singlesymptom.replace(' ', ' of ') + '"'
        
    if  (singlesymptom.find(' ') == singlesymptom.rfind(' ') and singlesymptom.find(' ') != -1):
        finalstring = finalstring + ',"' + singlesymptom.replace(' ', ' in ') + '"'
        
    return finalstring
```

```python
for singlesymptom in l:
    singlesymptom = singlesymptom.replace('>','').replace('(','').replace(')','')
    finalstring = singlesymptom
    finalstring = '"' + finalstring + '",' + '"' + finalstring + '"'
    if  (singlesymptom.find(' ') == singlesymptom.rfind(' ') and singlesymptom.find(' ') != -1):
        finalstring = getsimplevariants(singlesymptom, finalstring)
    if  (singlesymptom.find("Can't") != -1):
        finalstring = finalstring + ',"' + singlesymptom.replace("Can't","Cannot") + '"'
        finalstring = getsimplevariants(singlesymptom.replace("Can't","Cannot"), finalstring)
    if  (singlesymptom.find("ed") != -1):
        finalstring = finalstring + ',"' + singlesymptom.replace("ed","ment") + '"'
        finalstring = getsimplevariants(singlesymptom.replace("ed","ment"), finalstring)
    if  (singlesymptom.find("ed") != -1):
        finalstring = finalstring + ',"' + singlesymptom.replace("ed","ement") + '"'
        finalstring = getsimplevariants(singlesymptom.replace("ed","ement"), finalstring)
    if  (singlesymptom.find("ment") != -1):
        finalstring = finalstring + ',"' + singlesymptom.replace("ment","ed") + '"'
        finalstring = getsimplevariants(singlesymptom.replace("ment","ed"), finalstring)
    if  (singlesymptom.find("ed") != -1):
        finalstring = finalstring + ',"' + singlesymptom.replace("ement","ed") + '"'
        finalstring = getsimplevariants(singlesymptom.replace("ement","ed"), finalstring)
    if  (singlesymptom.find("ness") != -1):
        finalstring = finalstring + ',"' + singlesymptom.replace("ness","") + '"'
        finalstring = getsimplevariants(singlesymptom.replace("ness",""), finalstring)
    if  (singlesymptom.find("ing") != -1):
        finalstring = finalstring + ',"' + singlesymptom.replace("ing","y") + '"'
        finalstring = getsimplevariants(singlesymptom.replace("ing","y"), finalstring)
    if  (singlesymptom.find("pain") != -1):
        finalstring = finalstring + ',"' + singlesymptom.replace("pain","painful") + '"'
        finalstring = getsimplevariants(singlesymptom.replace("pain","painful"), finalstring)
    if  (singlesymptom.find("Pain") != -1):
        finalstring = finalstring + ',"' + singlesymptom.replace("Pain","Painful") + '"'
        finalstring = getsimplevariants(singlesymptom.replace("Pain","Painful"), finalstring)
    
    print(finalstring)
```
